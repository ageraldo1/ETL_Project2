```python
import requests
import json
from elasticsearch import Elasticsearch
import time
from datetime import date, timedelta
```

```python
es_url = 'http://192.168.99.100:9200'
index_name = 'theguardian'
api_key = ''

time_to_sleep = 10
reload = True 
```

```python
es = Elasticsearch(hosts=es_url)

if reload:
    es.indices.delete(index=index_name, ignore=[400, 404])
```

```python
index_name = 'tweets'

top_activity = {
  "size": 0, 
  "aggs": {
    "group_by_tweets": {
      "date_histogram": {
        "field": "timestamp",
        "interval": "month"
      },
      "aggs": {
        "total_tweets": {
          "value_count": {
            "field": "timestamp"
          }
        },
        "sales_bucket_sort" : {
          "bucket_sort": {
            "sort": [{"total_tweets" : {"order": "desc"} }],
            "size": 5
          }
          
        }
      }
    }
  }
}

res = es.search(index=index_name, body=top_activity)

search_dates = [record['key_as_string'] for record in res['aggregations']['group_by_tweets']['buckets']]
search_dates = ['2015-10-01']
```

```python
index_name = 'theguardian'
document_type = 'news'
page_number = 1

for dt in search_dates:
    while True: 
        params = {
            'api-key' : api_key,
            'from-date' : dt[0:10],
            'to-date' : dt[0:10],
            'page' : page_number
        }

        api_url = 'https://content.guardianapis.com/search'

        response = requests.get(url = api_url, params = params)
        #print((response.url))

        if response.status_code == 200:
            articles = response.json()

            if articles['response']['status'] == 'ok':
                for record in articles['response']['results']:
                    res = es.index(index=index_name, doc_type=document_type, body=record)

                    if 'failed' in res['result']:
                        print(f'Erros during tweets load : {res}')
                        break        

                time.sleep(time_to_sleep)
                page_number = page_number + 1

        elif response.status_code == 400:
            #requested page is beyond the number of available pages
            page_number = 1
            break

        else:
            print (f'Error loading api {url} ')
            break

print ('Process completed without issues')
```
